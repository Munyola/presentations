```python
%matplotlib notebook
import matplotlib.pyplot as plt
```

```python
style_cycle = [{'color':c, 'marker': m} for c, m in zip('rgb', 'xso')]

def pandas_plot(ax, df, style_cycle):
    """
    Plot a pandas DataFrame

    Parameters
    ----------
    ax : matplotlib.axes.Axes
        The axes to plot to

    df : pd.DataFrame
        The data to plot

    style_cycle : Cycler
        Something that when iterated over yields style dict
    """
    ret = {}
    x = df.index
    for n, sty in zip(df.columns, style_cycle):
        ln, = ax.plot(x, df[n], label=n, **sty)
        ret[n] = ln
    ax.legend()
    return ret
```

```python
plt.register_func(pandas_plot)
```

```python
import pandas as pd
import numpy as np
th = np.linspace(0, 2*np.pi, 128)
df = pd.DataFrame({'a': np.sin(th),
                   'b': np.sin(th + np.pi / 3), 
                   'c': np.sin(th + 2 * np.pi / 3)}, index=th)
                   
```

```python
plt.cla()
arts = plt.pandas_plot(df, style_cycle)
```
